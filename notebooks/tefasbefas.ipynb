```python
import pandas as pd, datetime, glob, os
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
```

Fon dağılımlarını yükleme

```python
csv_files = sorted(glob.glob(os.path.join("tefas2/dagilim", "*.csv")))
dag_dfs = {os.path.basename(f).split('.')[0]: pd.read_csv(f).fillna(0) for f in csv_files}
dates = [pd.to_datetime(key, format='%Y%m%d') for key in dag_dfs.keys()]
assets = list(dag_dfs[list(dag_dfs.keys())[0]].iloc[0].index)[3:]
```

fon verilerini indirme

```python
newdfs = {}
csv_files = sorted(glob.glob(os.path.join("tefas2/fonlar", "*.csv")))
fon_dfs = {os.path.basename(f).split('.')[0]: pd.read_csv(f).fillna(0) for f in csv_files}
columns_to_float = ['Fiyat','Tedavüldeki Pay Sayısı']
if len(fon_dfs.items()) != len(dag_dfs.items()):
    print('Fon dağılımları ve fiyat verileri dosyaları aynı sayıda değil!')
else: 
    for key, df in fon_dfs.items():
        newdf = df.merge(dag_dfs[key], on='Fon Kodu', how='left', suffixes=('', '_drop'))
        newdf = newdf.drop(columns=[col for col in newdf.columns if col.endswith('_drop')])
        newdf[columns_to_float+assets] = (newdf[columns_to_float+assets]
                                    .replace(r"\.", "", regex=True)
                                    .replace(",", ".", regex=True)
                                    .astype(float))
        #tefas_aum.append(newdf[columns_to_float+[myasset]].prod(axis=1).sum()/100)
        newdfs[key] = newdf
```
